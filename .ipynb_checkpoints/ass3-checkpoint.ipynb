```python
# Import necessary libraries
import numpy as np
import numpy.linalg as lg
import networkx as nx
import matplotlib.pyplot as plt
```

# ColoSpg network:

#### We start by loading the network `ColoSpg` from file, converts it to a standardized format, computes a layout, and generates a visualization - creating the foundation for all subsequent analysis. The adjacency matrix conversion is particularly important for spectral analysis later.

```python
# 1. Load the ColoSpg network
G_colo = nx.read_weighted_edgelist('Tortoise.txt', nodetype=int)  # Read edge list file with integer nodes

# 2. Create adjacency matrix
A_colo = nx.to_numpy_array(G_colo)  # Convert graph to numpy adjacency matrix

# 3. Redraw network with consistent node labeling
G_colo = nx.from_numpy_array(A_colo)  # Recreate graph from matrix for consistent ordering
pos_colo = nx.spring_layout(G_colo, seed=42)  # Calculate node positions using spring layout

# 4. Draw full network
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_colo, pos_colo, with_labels=False, node_size=20)
plt.title("ColoSpg Network")
plt.axis('off')
plt.savefig("ColoSpg_full.png")
plt.show()



```

#### Next, we calculates and displays each node's degree (number of connections) in the ColoSpg network using matrix operations

```python
# 5. Calculate basic properties
n_colo = len(A_colo)  # Number of nodes
e_colo = np.ones(n_colo)  # Vector of ones
k_colo = np.dot(A_colo, e_colo)  # Degree vector (sum of each row)
print("ColoSpg degree vector:\n", k_colo)
```

#### This following line performs spectral decomposition of the adjacency matrix to extract eigenvalues and eigenvectors for community detection.

```python
# 6. Eigenvalue decomposition
u_colo, v_colo = lg.eigh(A_colo)  # u=eigenvalues, v=eigenvectors
#print("u_colo =\n", u_colo, "\n")

#print("v_colo =\n", v_colo)

```

#### Here, we're Identifying and visuallying separates the network into two communities `C1_colo` and `C2_colo` using spectral partitioning.

```python
# 8. Community detection
P_colo = v_colo[:,n_colo-2]  # Eigenvector for most negative eigenvalue
C1_colo = np.where(P_colo > 0)[0]  # First community (positive components)
C2_colo = np.where(P_colo <= 0)[0] # Second community (non-positive)

# 9. Color nodes by community
color_colo = np.tile('g', n_colo)  # Default color green
color_colo[C1_colo] = 'b'          # Community 1 blue

# 10. Draw communities
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_colo, pos_colo, node_color=color_colo, with_labels=False, node_size=20)
plt.title("ColoSpg Communities")
plt.axis('off')
plt.savefig("ColoSpg_communities.png")
plt.show()
```

#### This following section quantifies and reports the community structure by calculating: internal edges within Community 1, internal edges within Community 2, crossing edges between the two communities, number of nodes in each community,...

```python
# 11. Calculate community edges
edges_within_colo_C_1 = np.sum(A_colo[np.ix_(C1_colo, C1_colo)])/2 
edges_within_colo_C_2 = np.sum(A_colo[np.ix_(C2_colo, C2_colo)])/2
edges_between_colo = np.sum(A_colo[np.ix_(C1_colo, C2_colo)])
total_edges_colo = np.sum(A_colo)/2

print("\nCommunity Structure:\n")

print(f"Community 1 size: {len(C1_colo)}, Community 2 size: {len(C2_colo)}\n")

print("Edges Within C1:", edges_within_colo_C_1, "\n")

print("Edges Within C2:", edges_within_colo_C_2, "\n")

print("Edges between C1 and C2:", edges_between_colo, "\n")
      
print("Total:", total_edges_colo)
```

#### This part of the code calculates and reports key network properties clustering coefficient (local connectivity) and motif counts (triangles C3, squares C4) to quantify structural patterns and compare them against random networks, revealing organizational principles of the ColoSpg network.

```python
# 7. Count triangles and squares
triangles_colo = sum(u_colo**3)/6  # Tr(A^3) = 6*number of triangles

# Calculate |P₁| (number of edges)
m = G_colo.number_of_edges()

# Calculate |P₂| (number of paths of length 2)
degrees = [d for n, d in G_colo.degree()]
P2 = sum(d*(d-1)/2 for d in degrees)  # sum of (k_i choose 2)

# Calculate μ₄ (sum of fourth powers of eigenvalues)
mu4 = sum(u_colo**4)

# Now calculate C4 using the correct formula
squares_colo = (mu4 - 2*m - 4*P2)/8

# Calculate clustering coefficient and motifs
ws_clustering = nx.average_clustering(G_colo)

print("\nClustering and Motifs:\n")
print(f"Watts-Strogatz clustering coefficient:", ws_clustering, "\n")
print("ColoSpg triangles (C3):", triangles_colo, "\n")
print("ColoSpg squares (C4):", squares_colo)
```

#### Here we're identifying the most influential nodes in the network by computing and ranking three centrality measures (degree, closeness, betweenness) to reveal key structural positions—highlighting nodes that are highly connected, centrally located, or critical for information flow.

```python
# Centrality measures
degree_centrality = nx.degree_centrality(G_colo)
closeness_centrality = nx.closeness_centrality(G_colo)
betweenness_centrality = nx.betweenness_centrality(G_colo)

# Get top 5 nodes for each centrality
top_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
top_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
top_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

print("\nTop Central Nodes:\n")
print("Degree centrality:", [x[0] for x in top_degree], "\n")
print("Closeness centrality:", [x[0] for x in top_closeness], "\n")
print("Betweenness centrality:", [x[0] for x in top_betweenness])
```

#### This part of the code creates visualizations of the network's most central nodes by highlighting the top 5 nodes for each centrality measure (degree, closeness, betweenness) to reveal their structural importance and spatial distribution within the ColoSpg network.

```python
# 13. Visualize top nodes for each centrality measure (default colors with highlighted top nodes)

# Define a color palette for top nodes
top_colors = ['#FF0000', '#00FF00', '#0000FF', '#FF00FF', '#00FFFF']  # Red, Green, Blue, Magenta, Cyan

# Degree centrality visualization
plt.figure(figsize=(12, 10))
# Draw the base network with default colors
nx.draw_networkx(G_colo, pos_colo, with_labels=False, node_size=20, edge_color='gray')

# Highlight top degree centrality nodes
for i, (node, _) in enumerate(top_degree):
    nx.draw_networkx_nodes(G_colo, pos_colo, nodelist=[node], 
                         node_color=top_colors[i], node_size=100+(4-i)*20)
    
plt.title("ColoSpg Network - Top 5 Degree Centrality Nodes")
plt.axis('off')
plt.savefig("ColoSpg_degree_centrality.png")
plt.show()
```

```python


# Closeness centrality visualization
plt.figure(figsize=(12, 10))
# Draw the base network with default colors
nx.draw_networkx(G_colo, pos_colo, with_labels=False, node_size=20, edge_color='gray')

# Highlight top closeness centrality nodes
for i, (node, _) in enumerate(top_closeness):
    nx.draw_networkx_nodes(G_colo, pos_colo, nodelist=[node], 
                         node_color=top_colors[i], node_size=100+(4-i)*20)
    
plt.title("ColoSpg Network - Top 5 Closeness Centrality Nodes")
plt.axis('off')
plt.savefig("ColoSpg_closeness_centrality.png")
plt.show()

```

```python
# Betweenness centrality visualization
plt.figure(figsize=(12, 10))
# Draw the base network with default colors
nx.draw_networkx(G_colo, pos_colo, with_labels=False, node_size=20, edge_color='gray')

# Highlight top betweenness centrality nodes
for i, (node, _) in enumerate(top_betweenness):
    nx.draw_networkx_nodes(G_colo, pos_colo, nodelist=[node], 
                         node_color=top_colors[i], node_size=100+(4-i)*20)
    
plt.title("ColoSpg Network - Top 5 Betweenness Centrality Nodes")
plt.axis('off')
plt.savefig("ColoSpg_betweenness_centrality.png")
plt.show()
```

#### This code generates and visualizes a random Erdős-Rényi graph with the same node count and expected edge density as the ColoSpg network, enabling structural comparisons between real-world and random networks

```python
# 12. Create and draw random graph comparison
m_colo = total_edges_colo
p_colo = 2*m_colo/(n_colo*(n_colo-1))  # Connection probability
G_random_colo = nx.erdos_renyi_graph(n_colo, p_colo, seed=42)  # Add seed for reproducibility

# Draw the random graph
plt.figure(figsize=(12, 10))
pos_random_colo = nx.spring_layout(G_random_colo, seed=42)  # Consistent layout
nx.draw_networkx(G_random_colo, pos_random_colo, with_labels=False, node_size=40, alpha=0.7)
plt.title(f"Random Graph Comparison\n(n={n_colo}, p={p_colo:.6f}, edges={G_random_colo.number_of_edges()})")
plt.axis('off')
plt.tight_layout()
plt.savefig("ColoSpg_random_comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Here we compute the adjacency matrix and eigenvalues of the random graph created, for comparative analysis against the real network.

```python
# 12. Create random graph comparison
m_colo = total_edges_colo
G_random_colo = nx.erdos_renyi_graph(n_colo, 2*m_colo/(n_colo*(n_colo-1)))
A_random_colo = nx.to_numpy_array(G_random_colo)
u_random_colo, _ = lg.eigh(A_random_colo)
```

#### This part calculates and compares clustering coefficients and motif counts (triangles C3, squares C4) between the real ColoSpg network and its random counterpart to identify statistically significant structural patterns.

```python
# 14. Count triangles and squares in the random network
random_triangles_colo = sum(u_random_colo**3)/6  # Tr(A^3) = 6*number of triangles

# Calculate |P₁| (number of edges in the random network)
random_m = G_random_colo.number_of_edges()

# Calculate |P₂| (number of paths of length 2 in the random network)
random_degrees = [d for n, d in G_random_colo.degree()]
random_P2 = sum(d*(d-1)/2 for d in random_degrees)  # sum of (k_i choose 2)

# Calculate μ₄ (sum of fourth powers of eigenvalues)
random_mu4 = sum(u_random_colo**4)

# Now calculate C4 using the correct formula
random_squares_colo = (random_mu4 - 2*random_m - 4*random_P2)/8

# Calculate clustering coefficient and motifs
random_ws_clustering = nx.average_clustering(G_random_colo)

print("\nClustering and Motifs:\n")
print(f"Random Watts-Strogatz clustering coefficient:", random_ws_clustering, "\n")
print("Random ColoSpg network triangles (C3):", random_triangles_colo, "\n")
print("Random ColoSpg squares (C4):", random_squares_colo)
```

#### And then, we calculate and compare centrality measures (degree, closeness, betweenness) between the real ColoSpg network and its random counterpart to identify statistically significant node importance patterns.

```python
# Centrality measures
degree_centrality = nx.degree_centrality(G_random_colo)
closeness_centrality = nx.closeness_centrality(G_random_colo)
betweenness_centrality = nx.betweenness_centrality(G_random_colo)

# Get top 5 nodes for each centrality
random_top_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
random_top_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
random_top_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

print("\nRandom Top Central Nodes:\n")
print("Random Degree centrality:", [x[0] for x in random_top_degree], "\n")
print("Random Closeness centrality:", [x[0] for x in random_top_closeness], "\n")
print("Random Betweenness centrality:", [x[0] for x in random_top_betweenness])
```

# Tortoise network:

#### We follow exactly the same prcedure as before.

```python
# 1. Load the Tortoise network
G_tortoise = nx.read_weighted_edgelist('Tortoise.txt', nodetype=int)

# 2. Create adjacency matrix
A_tortoise = nx.to_numpy_array(G_tortoise)

# 3. Redraw network with consistent node labeling
G_tortoise = nx.from_numpy_array(A_tortoise)
pos_tortoise = nx.spring_layout(G_tortoise, seed=42)

# 4. Draw full network
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_tortoise, pos_tortoise, with_labels=True, node_size=500)
plt.title("Tortoise Network")
plt.axis('off')
plt.savefig("Tortoise_full.png")
plt.show()

```

```python
# 5. Calculate basic properties
n_tortoise = len(A_tortoise)
e_tortoise = np.ones(n_tortoise)
k_tortoise = np.dot(A_tortoise, e_tortoise)
print("Tortoise degree vector:\n", k_tortoise)

```

```python
# 6. Eigenvalue decomposition
u_tortoise, v_tortoise = lg.eigh(A_tortoise)

# 7. Count triangles and squares
triangles_tortoise = sum(u_tortoise**3)/6  # Tr(A^3) = 6*number of triangles

# Calculate |P₁| (number of edges in the tortoise network)
m_tortoise = G_tortoise.number_of_edges()

# Calculate |P₂| (number of paths of length 2 in the tortoise network)
degrees_tortoise = [d for n, d in G_tortoise.degree()]
P2_tortoise = sum(d*(d-1)/2 for d in degrees_tortoise)  # sum of (k_i choose 2)

# Calculate μ₄ (sum of fourth powers of eigenvalues)
mu4_tortoise = sum(u_tortoise**4)

# Now calculate C4 using the correct formula
squares_tortoise = (mu4_tortoise - 2*m_tortoise - 4*P2_tortoise)/8

# Calculate clustering coefficient and motifs
print("\nClustering and Motifs:")
print(f"Watts-Strogatz clustering coefficient:", nx.average_clustering(G_tortoise))
print("Tortoise triangles (C3):", triangles_tortoise)
print("Tortoise squares (C4):", squares_tortoise)
```

```python
# 8. Community detection
P_tortoise = v_tortoise[:,n_tortoise-2]  # Eigenvector for second largest eigenvalue
C1_tortoise = np.where(P_tortoise > 0)[0]
C2_tortoise = np.where(P_tortoise <= 0)[0]

# 9. Color nodes by community
color_tortoise = np.tile('g', n_tortoise)  # Default green
color_tortoise[C1_tortoise] = 'b'          # Community 1 blue

# 10. Draw communities
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_tortoise, pos_tortoise, node_color=color_tortoise, with_labels=True, node_size=500)
plt.title("Tortoise Communities")
plt.axis('off')
plt.savefig("Tortoise_communities.png")
plt.show()
```

```python

# 11. Calculate community edges
edges_within_C1 = np.sum(A_tortoise[np.ix_(C1_tortoise, C1_tortoise)])/2 
edges_within_C2 = np.sum(A_tortoise[np.ix_(C2_tortoise, C2_tortoise)])/2
edges_between = np.sum(A_tortoise[np.ix_(C1_tortoise, C2_tortoise)])
total_edges = np.sum(A_tortoise)/2

print("\nCommunity Structure:")
print(f"Community 1 size: {len(C1_tortoise)}, Community 2 size: {len(C2_tortoise)}")
print(f"Edges within C1: {edges_within_C1}, within C2: {edges_within_C2}, between: {edges_between}")
print(f"Total edges: {total_edges}")

```

```python
# 12. Centrality measures
degree_centrality = nx.degree_centrality(G_tortoise)
closeness_centrality = nx.closeness_centrality(G_tortoise)
betweenness_centrality = nx.betweenness_centrality(G_tortoise)

# Get top 5 nodes for each centrality
top_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
top_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
top_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

print("\nTop Central Nodes:")
print("Degree centrality:", [x[0] for x in top_degree])
print("Closeness centrality:", [x[0] for x in top_closeness])
print("Betweenness centrality:", [x[0] for x in top_betweenness])

```

```python
# 13. Visualize top nodes for each centrality measure
top_colors = ['#FF0000', '#00FF00', '#0000FF', '#FF00FF', '#00FFFF']  # Red, Green, Blue, Magenta, Cyan

# Degree centrality visualization
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_tortoise, pos_tortoise, with_labels=True, node_size=500, edge_color='gray')
for i, (node, _) in enumerate(top_degree):
    nx.draw_networkx_nodes(G_tortoise, pos_tortoise, nodelist=[node], 
                         node_color=top_colors[i], node_size=700+(4-i)*50)
plt.title("Tortoise Network - Top 5 Degree Centrality Nodes")
plt.axis('off')
plt.savefig("Tortoise_degree_centrality.png")
plt.show()


```

```python

# Closeness centrality visualization
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_tortoise, pos_tortoise, with_labels=True, node_size=500, edge_color='gray')
for i, (node, _) in enumerate(top_closeness):
    nx.draw_networkx_nodes(G_tortoise, pos_tortoise, nodelist=[node], 
                         node_color=top_colors[i], node_size=700+(4-i)*50)
plt.title("Tortoise Network - Top 5 Closeness Centrality Nodes")
plt.axis('off')
plt.savefig("Tortoise_closeness_centrality.png")
plt.show()
```

```python

# Betweenness centrality visualization
plt.figure(figsize=(10, 8))
nx.draw_networkx(G_tortoise, pos_tortoise, with_labels=True, node_size=500, edge_color='gray')
for i, (node, _) in enumerate(top_betweenness):
    nx.draw_networkx_nodes(G_tortoise, pos_tortoise, nodelist=[node], 
                         node_color=top_colors[i], node_size=700+(4-i)*50)
plt.title("Tortoise Network - Top 5 Betweenness Centrality Nodes")
plt.axis('off')
plt.savefig("Tortoise_betweenness_centrality.png")
plt.show()
```

```python
# 14. Create random graph comparison
m_tortoise = total_edges
G_random_tortoise = nx.erdos_renyi_graph(n_tortoise, 2*m_tortoise/(n_tortoise*(n_tortoise-1)))
A_random_tortoise = nx.to_numpy_array(G_random_tortoise)
p_tortoise = 2*m_tortoise/(n_tortoise*(n_tortoise-1))  # Connection probability

u_random_tortoise, _ = lg.eigh(A_random_tortoise) #eigenvalues

# Draw the random graph
plt.figure(figsize=(10, 8))
pos_random = nx.spring_layout(G_random_tortoise, seed=42)  # Consistent layout for comparison
nx.draw_networkx(G_random_tortoise, pos_random, with_labels=True, node_size=500, node_color='lightblue')
plt.title(f"Random Graph (n={n_tortoise}, p={p_tortoise:.3f})")
plt.axis('off')
plt.savefig("Tortoise_random_graph.png")
plt.show()
```

```python
# 15. Random network analysis
random_triangles = sum(u_random_tortoise**3)/6  # Correct formula for triangles

# Calculate components for C4 formula
random_m = G_random_tortoise.number_of_edges()  # |P₁| (number of edges)
random_degrees = [d for n, d in G_random_tortoise.degree()]  # Node degrees
random_P2 = sum(d*(d-1)/2 for d in random_degrees)  # |P₂| (paths of length 2)
random_mu4 = sum(u_random_tortoise**4)  # μ₄ (sum of eigenvalue^4)

# Correct formula for squares (C4)
random_squares = (random_mu4 - 2*random_m - 4*random_P2)/8

print("\nRandom Network Comparison:")
print(f"Random Watts-Strogatz clustering coefficient:", nx.average_clustering(G_random_tortoise))
print("Random triangles (C3):", random_triangles)
print("Random squares (C4):", random_squares)
```
